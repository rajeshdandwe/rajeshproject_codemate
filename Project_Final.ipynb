**About dataset:**
The Sleep Health and Lifestyle Dataset comprises 400 rows and 13 columns, covering a wide range of variables related to sleep and daily habits. It includes details such as gender, age, occupation, sleep duration, quality of sleep, physical activity level, stress levels, BMI category, blood pressure, heart rate, daily steps, and the presence or absence of sleep disorders.

Comprehensive Sleep Metrics: Explore sleep duration, quality, and factors influencing sleep patterns.
Lifestyle Factors: Analyze physical activity levels, stress levels, and BMI categories.
Cardiovascular Health: Examine blood pressure and heart rate measurements.        
Sleep Disorder Analysis: Identify the occurrence of sleep disorders such as Insomnia and Sleep Apnea.

**Details about Sleep Disorder Column:**  
**None:** The individual does not exhibit any specific sleep disorder.  
**Insomnia:** The individual experiences difficulty falling asleep or staying asleep, leading to inadequate or poor-quality sleep.  
**Sleep Apnea:** The individual suffers from pauses in breathing during sleep, resulting in disrupted sleep patterns and potential health risks.

# **Train Test Split**

```python
x=df.drop('Sleep Disorder',axis=1)
x
```

```python
y=df['Sleep Disorder']
y
```

In x variable split all column except sleep disorder in y we store a sleep disorder column


**Logistic regression**

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
```

```python
print("Accuracy Score:",round(accuracy_score(ytest,ypred)*100,2),"%")
```

```python
print(classification_report(ytest,ypred))
```

```python
print(confusion_matrix(ytest,ypred))
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

**Prediction Of Unknown value wheather a person have Sleeping Disorder or Not:**

```python
ge=float(input("Enter gender O for female 1 fro male"))
age=int(input("Enter the Age:"))
occu=float(input("Enter Occupation[0:SoftwareEng,1:Doctor,2:Sales Representative,3:Teacher,4:Nurse,5:Eng,6:Accountant,7:Scientist,8:Lawyer,9:salesperson,10:Manager]"))
sleepd=float(input("Enter the Sleep Duration:"))
qos=int(input("Enter Quality of Sleep:"))
pal=int(input("Enter Physical Activity Level(20-100):"))
st=int(input("Enter Stress Level(1-10):"))
bmi=float(input("Enter BMI Category between 0 to 2:"))
hr=int(input("Enter Heart rate:"))
ds=int(input("Enter Daily Sleep Second(1000-10000):"))
bp1=int(input("Enter Blood Pressure 1:"))
bp2=int(input("Enter Blood Pressure 2:"))
unknown_x=[[ge,age,occu,sleepd,qos,pal,st,bmi,hr,ds,bp1,bp2]]
```

```python
prediction=model.predict(unknown_x)
if(prediction==1.0):
    print("None")
elif(prediction==2.0):
    print("Sleep Apenia")
elif(prediction==0.0):
  print("Insomnia")
```
